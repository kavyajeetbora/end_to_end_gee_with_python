## Drawing Tools in **geemap**



```python
Map = geemap.Map(center=[40,-100], zoom=5)

# Add Earth Engine dataset
image = ee.Image('USGS/SRTMGL1_003')

# Set visualization parameters.
vis_params = {
    'min': 0,
    'max': 4000,
    'palette': ['006633', 'E5FFCC', '662A00', 'D8D8D8', 'F5F5F5'],
}

# Add Earth Engine DEM to map
Map.addLayer(image, vis_params, 'SRTM DEM')

states = ee.FeatureCollection("TIGER/2018/States")  ## Feature collection containing all states polygons
Map.addLayer(states, {}, 'US States')

Map
```

## Retrieving the objects from the above map

```python
## List all objects
Map.draw_features
```

```python
## Select the last feature
Map.draw_last_feature
```

Based on the geometry objects drawn on the map, filter out the states that intersects these drawn geometries:

```python
roi = ee.FeatureCollection(Map.draw_features)
selected_states = states.filterBounds(roi)
Map.addLayer(selected_states, {}, 'Selected States')
```

Clip the DEM raster from the selected state polygons:

```python
clipped_image =  image.clip(selected_states)
Map.addLayer(clipped_image, vis_params, "Clipped DEM")
Map
```
