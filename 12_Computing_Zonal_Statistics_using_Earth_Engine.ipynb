## Add Earth engine dataset


Loading SRTM image - Shuttle Radar Topography Mission

Captures the digital elevation data with spatial resolution = 30m

```python
image = ee.Image('USGS/SRTMGL1_003')
image
```

Set the visualization parameter and plot

```python
Map = geemap.Map(center=[21.79, 70.87], zoom=3)

## DEM from Earth Engine
dem = ee.Image('USGS/SRTMGL1_003')
vis_params = {
    'min': 0,
    'max': 4000,
    'palette': ["006633", "E5FFCC", "662A00", "D8D8D8", "F5F5F5"],
}
## Add the layer to map
Map.addLayer(ee_object = dem, vis_params = vis_params, name = 'SRTM DEM', shown=True, opacity=0.7)


## Lansat Data from Earth Engine
landsat = ee.Image("LANDSAT/LE7_TOA_5YEAR/1999_2003")
landsat_vis = {"bands":['B4','B3','B2'], "gamma":1.4}
Map.addLayer(landsat, landsat_vis, 'LANDSAT/LE7_TOA_5YEAR/1999_2003')

## Add State level administration boundary from google earth feature collection
states = ee.FeatureCollection("TIGER/2018/States")
Map.addLayer(states, {}, 'US States')
Map
```

## Compute the zonal statistics


Computing the average of DEM values & Landsat values for each state and exporting it to a csv file

```python
dem_stats = geemap.zonal_stats(
    in_value_raster = dem,
    in_zone_vector=states,
    out_file_path='stats.csv',
    stat_type='MEAN',
    scale=1000
)
```

```python
landsat_stats = geemap.zonal_stats(
    in_value_raster = landsat,
    in_zone_vector=states,
    out_file_path='landsat.csv',
    stat_type='SUM',
    scale=1000
)
```
